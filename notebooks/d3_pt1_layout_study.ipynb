## D3 plot size and margins
The plot margins are stored in an object. Next, the width and height are declared in a way that we can conceptually think of them as the entire drawing area of the plot _minus_ the margins. 

This is known as the ['margin convention'](https://bl.ocks.org/mbostock/3019563)
```
// .js code
var margin = { top: 20, right: 30, bottom: 30, left: 40 },
width = 800 - margin.left - margin.right,
height = 500 - margin.top - margin.bottom;
```

*[Observable's version of margin convention](https://observablehq.com/@d3/margin-convention)

## `svg` element
The entire plot will be drawn within an `svg` elmement. This element is scaled to the `width=800` and `height=500`.  Notice again that we use the relationship of the width and height to the margins by adding them when we set the respective attributes.  

This practice enables us to set the height & width margins according to our original size and margin specification in one location and then allows us to refer to it throughout the visualization without explicitly setting the values `800` and `500` _i.e.,_ avoid using "magic numbers".
```
// .js code
var svg = d3.select("body")
    .append("svg")
    .attr("id", "container")
    .attr("width", width + margin.left + margin.right)
    .attr("height", height + margin.bottom + margin.top);
```
(The retrieved dimensions of the `svg` account for a 1 px border)

````python
svg = driver.find_element_by_id("container")
svg_src = svg.get_attribute("outerHTML")
svg_transform = svg.value_of_css_property("transform")
print(f"svg dims H:{svg.size['height']} X W:{svg.size['width']}")
print(f"svg transform: {svg_transform}")
````

## `g` element layout and sizing
A grouping `g` element is used to hold the plot elements.  Since the origin is oriented at the top-left, this element is translated using the top and left margin values.
```
// .js code
var plot = svg.append("g")
    .attr("id","plot")
    .attr("transform", "translate("+margin.left+","+margin.top+")");
```
The translation of the `g` element can be retrieved using a CSS `transform` lookup.  The last two values in the matrix represent the `xy` transform.

````python
g = driver.find_element_by_id("plot")
g_transform = g.value_of_css_property("transform")
print(f"group transform: {g_transform}")
````

To demonstrate the plot area within the `g` nested in the `svg`, we insert a `rect` element.
```
//.js code
var plot_area = plot.append("rect")
    .attr("id", "plot_area")
    .attr("width", width)
    .attr("height", height);
```

### Scale Setup

For our study, we add two linear scales for the `x` and `y` axis.  Note that we can now ignore handling the margins and simply declare the ranges as follows:

```
//.js code
var x = d3.scaleLinear()
    .domain([0,1])
    .range([0, width]);

var y = d3.scaleLinear()
    .domain([0,1])
    .range([height, 0]);  
```
Both axes are added to the `<g id=plot>` element.
```
//.js code
var x_axis = plot.append("g")
    .attr("id", "x_axis")
    .attr("class", "axis")
    .attr("transform", "translate(0," + height + ")")            
    .call(d3.axisBottom(x));

var y_axis = plot.append("g")
    .attr("id", "y_axis")
    .attr("class", "axis")
    .call(d3.axisLeft(y)); 
```

````python
url = 'http://localhost:3000/margin_study_scales.html'
driver.get(url)
rendered_width = driver.execute_script("return document.documentElement.scrollWidth")
rendered_height = driver.execute_script("return document.documentElement.scrollHeight")
svg = driver.find_element_by_id("container")
driver.set_window_size(svg.size['width']+10,svg.size['height']+20)
svg_coords = svg.location_once_scrolled_into_view 
driver.execute_script('window.scrollTo({}, {});'.format(svg_coords['x'], svg_coords['y']))
driver.save_screenshot("img/scales.png")
driver.set_window_size(rendered_width, rendered_height)
plt.rcParams['figure.dpi'] = 300
img = mpimg.imread('img/scales.png')
fig, ax = plt.subplots()
ax.set_axis_off()
imgplot = plt.imshow(img)
````

### Scale Comprehension
After defining our scales and adding our axes, we use Selenium to obtain information about a scale.

We know that our `x` scale was defined to have a range from `0`-`width` and placed at the bottom of the plot. In Selenium, we get the `x_axis` group to comprehend these attributes.  We look at a special `path` element with the `class=domain` that is created when we inserted the axis.  The size of this path represents the domain of `x` scale function sized to the range we specified. 

#### Scale Width

````python
x_axis_domain = driver.find_element_by_id("x_axis").find_element_by_tag_name("path")
x_axis_domain_src = x_axis_domain.get_attribute("outerHTML")
print(f"path element src: {x_axis_domain_src}")
print(f"x-axis width: {x_axis_domain.size['width']}")
````

#### Tick Labels
We can further exploit the group structure of an axis to find the amount of tick marks and the tick marks text labels. 

````python
x_axis_tick_groups = driver.find_element_by_id("x_axis").find_elements_by_class_name("tick")
x_axis_ticks_labels = [t.find_element_by_tag_name("text").text for t in  x_axis_tick_groups]
print(f"{len(x_axis_tick_groups)} x-axis ticks")
print(f"tick labels: {x_axis_ticks_labels}")
````

#### Scale / Tick Mark Orientation

The `x` axis was created with the `d3.axisBottom(x)` call.   In this orientation, ticks are drawn [below](https://github.com/d3/d3-axis#axisBottom) the horizontal 
domain path.  To verify orientation, we study the `y2` positions of each tick mark (line element).

````python
x_tick_marks = [g.find_elements_by_tag_name("line") for g in x_axis_tick_groups]
tick = x_tick_marks[0][0].get_attribute("outerHTML")
print(f"tick src: {tick}")
x_tick_marks_y2 = [np.int(mark[0].get_attribute('y2')) for mark in x_tick_marks] #orientation of tick mark
print(x_tick_marks_y2)
````

Positive `y2` values indicate that the tick marks extend downward and below the `x` axis.

We can verify the left orientation of the `y` axis in a similar manner.

````python
y_tick_groups = driver.find_element_by_id("y_axis").find_elements_by_class_name("tick")
y_tick_marks = [g.find_elements_by_tag_name("line") for g in y_tick_groups]
y_tick = y_tick_marks[0][0].get_attribute("outerHTML")
print(f"tick src: {y_tick}")
y_tick_marks_x2 = [np.int(mark[0].get_attribute('x2')) for mark in y_tick_marks] #orientation of tick mark
print(y_tick_marks_x2)
````

Negative `x2` values indicate that the tick marks extend outward and left of the `y` axis.
